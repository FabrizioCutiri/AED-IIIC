```python
%%writefile Ejercicio1.cpp
int s = 0;                             //1 inicialización
double prom = 0;                       //1 inicialización
for (int i = 0; i < n; i++) {          //1 + n(1+3+3+2) = 9n + 1
  s = s + A[i];                           // 3
  prom = s / (double) n;                  // 3
}
printf("Suma:%d\nProm:%d", s, prom);   //2

//Tiempo Detallado: 1 + 1 + 9n + 1 + 2 = 9n + 5
//Tiempo Asintótico: O(n) 
```

```python
%%writefile Ejercicio3.cpp
for (int i = 0; i < n - 1; i++) {           // 1 + (n-1)(2 + 13n-11 + 2) = 1+(n-1)(13n-7) = 1+13n^2-20n+7 = 13n^2-20n+8
  for (int k = i + 1; k < n; k++) {             //2 + (n-1)(1 +10+2) = 13n-11
    if (vec[i] > vec[k]) {                          //3
      int aux = vec[i];                             //2
      vec[i] = vec[k];                              //3
      vec[k] = aux;                                 //2
    }
  }
}

//Tiempo Detallado = 13n^2- 20n + 8
//Tiempo Asintótico = O(n^2)
```

```python
%%writefile Ejercicio4.cpp
int frec[101] = {0};                      //1 inicialización
for (int i = 0; i < n; i++) {             // 1 + n(1+5+2)
  frec[vec[i]]++;                             //5
}
int pos = 0;                              //1 inicialización
for (int i = 0; i < 101; i++) {           //1 + 101(1+)
    for (int k = 0; k < frec[i]; k++) {       //1 + 
      vec[pos] = i;                               //2
      pos++;                                      //2
    }
}
```

```python
%%writefile Ejercicio5.cpp
int i=0;                                //1 Inicializacion
int sum=0;                              //1 Inicializacion
while (i < 100) {                       // 100(6n + 3)  ----> Nota: De las dos condicionales se selecciona el de mayor complejidad
//--------------Condicional 1------------------------------(6n+3)
  if (i % 2 == 0) {                         //2
    for (int k = 0; k < n; k++) {           //1 + n(1+3+2)
    sum += vec[i];                              //3
    }
  }
//-------------Condicional 2 -------------------------------(601)
  else {
    for (int k=0; k<i; k++) {              //1 + 100(1+3+2)
      sum += vec[i];                          //3
    }
  }
}

//-------------------Resultado----------------------------
//Tiempo Detallado:  1 + 1 + 100(6n + 3) = 600n + 302
//Tiempo Asintótico: O(n)
```

```python
%%writefile Ejercicio8.cpp

int pos = 0;                          //1 Inicialización
for (int i = 1; i < n; i++) {         //1 + (n-1)(1+4+2)
  if (vec[i] > vec[pos]){                 //3
    pos = i;                              //1 Asignación
  }
}
printf("El mayor es: %d", vec[pos]);  //2

//-------------------Resultado----------------------------
//Tiempo Detallado:  1 + 1+ (n-1)(1+4+2) = 7n - 5
//Tiempo Asintótico:  O(n)
```

```python
%%writefile Ejercicio9.cpp
int max = 0;                          //1 Inicialización
for (int m = 0; m < n; m++) {         //1 + n(1+1+3+n(4+2+2+1+1)+2)= 1+ n(8n+6) = 8n^2+6n+1
    int cont = 0;                         //1 Inicialización
    int k = m + 1;                        //3
    while (vec[m] <= vec[k]) {            //n(4+2+2)
        k = k + 1;                          //2
        cont++;                             //2
    }
    if (cont > max){                      //1
      max = cont;                         //1
    }
}
printf("Maximo %d", max);             //1

//-------------------Resultado----------------------------
//Tiempo Detallado:  1 + 8n^2+6n+1 +1 = 8n^2+6n+3
//Tiempo Asintótico:  O(n^2)
```
